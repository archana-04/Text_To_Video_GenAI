```python
news = """Once in a small town, there was a little girl named Mia who loved flying kites.
One windy day, Mia's favorite kite soared higher and higher until it slipped from her grasp, disappearing into the vast sky.
Heartbroken, Mia decided to embark on a journey to find her lost kite.
As she roamed through the town, she encountered helpful neighbors and friendly animals, each offering a clue to the kite's whereabouts.
With determination in her heart, Mia followed the clues, overcoming obstacles and making new friends along the way.
Eventually, she reached a hill where her kite had become entangled in a tree.
Mia carefully climbed the tree, rescued her beloved kite, and rejoiced in the triumph of their reunion.
Grateful for the kindness she experienced on her journey, Mia returned to her town, sharing stories of adventure and the joy of never giving up.
The townsfolk celebrated Mia's bravery, and from that day forward, every windy day became an opportunity for the community to come together and fly kites, creating colorful memories in the sky."""
```

```python
import re
news_modules = news.split("\n")
```

```python
news_modules = [string for string in news_modules if string!=" "]
print(news_modules)

for prompt in news_modules:
  video_frames = pipe(prompt, num_inference_steps=40, height=320, width=576, num_frames=25).frames
  video_path = export_to_video(video_frames)
  print(type(video_path))
  print(video_path)
  videos.append(VideoFileClip(video_path))
  print(type(videos[-1]))

finalvid = concatenate_videoclips(videos,method="compose")
finalvid.write_videofile("final.mp4")


```
